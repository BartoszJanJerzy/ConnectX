# Create ConnectX Environment

```python
from kaggle_environments import evaluate, make, utils
import numpy as np

env = make("connectx", debug=True)
env.render()
```

# Example agents

```python
def random_agent(observation, configuration) -> int:
    '''Random agent'''
    from random import choice
    return choice([c for c in range(configuration.columns) if observation.board[c] == 0])
```

```python
def agent_where_is_max(observation, configuration) -> int:
    '''Chooses col with maximum of checkers.'''
    
    columns = configuration.columns # Number of Columns on the Board.
    rows = configuration.rows # Number of Rows on the Board.
    inarow = configuration.inarow # Number of Checkers "in a row" needed to win.
    board = observation.board # The current serialized Board (rows x columns).
    mark = observation.mark # Which player the agent is playing as (1 or 2).
    
    r_board = np.array(board).reshape(6, 7).T
    final_col = int(np.random.choice([0,1,2,3,4,5,6]))

    sums = []
    for col in [0,1,2,3,4,5,6]:
        current_sum = sum([int(bool(x)) for x in r_board[col]])
        sums.append(current_sum)
    
    for i in range(len(sums)):
        if sums[i] == max(sums) and sums[i] != 7:
            final_col = i
            break

    return final_col
```

```python
def agent_where_is_min(observation, configuration) -> int:
    '''Chooses col with min of checkers.'''

    columns = configuration.columns # Number of Columns on the Board.
    rows = configuration.rows # Number of Rows on the Board.
    inarow = configuration.inarow # Number of Checkers "in a row" needed to win.
    board = observation.board # The current serialized Board (rows x columns).
    mark = observation.mark # Which player the agent is playing as (1 or 2).
    
    r_board = np.array(board).reshape(6, 7).T
    final_col = int(np.random.choice([0,1,2,3,4,5,6]))

    sums = []
    for col in [0,1,2,3,4,5,6]:
        current_sum = sum([int(bool(x)) for x in r_board[col]])
        sums.append(current_sum)
    
    for i in range(len(sums)):
        if sums[i] == min(sums):
            final_col = i
            break

    return final_col
```

# Test your Agent

```python
env.reset()
# Play as the first agent against default "random" agent.
env.run([agent_where_is_min, "random"])
env.render(mode="ipython", width=1000, height=450)
```
